```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv('/content/TAM.csv')
# Convertir 'fecha' a datetime, especificando el formato '%Y-%m'
df['fecha'] = pd.to_datetime(df['fecha'], format='%Y-%m', errors='coerce')
# The errors='coerce' argument will handle any invalid dates by setting them to NaT (Not a Time)

# 1. Tamaño del Mercado Geográfico
geo_group = df.groupby('Estado')['per_ocu'].sum().reset_index()
print("Tamaño del mercado por estado:")
print(geo_group)

# 2. Segmentación por Clasificación Empresarial
clasif_group = df.groupby('clasificacion_empresa')['per_ocu'].sum().reset_index()
print("\nDistribución de personal ocupado por clasificación empresarial:")
print(clasif_group)

# 3. Potencial de Mercado por Actividad Económica
actividad_group = df.groupby('nombre_act')['per_ocu'].sum().reset_index()
print("\nPotencial de mercado por actividad económica:")
print(actividad_group)

# 4. Análisis Temporal
# Group by month and sum 'per_ocu'
temporal_group = df.groupby(df['fecha'].dt.to_period('M'))['per_ocu'].sum().reset_index()
# Convert the 'fecha' column back to string for plotting
temporal_group['fecha'] = temporal_group['fecha'].astype(str)
print("\nEvolución del mercado a lo largo del tiempo:")
print(temporal_group)
```

```python
# Importar las bibliotecas necesarias
import pandas as pd
from pyspark.sql import SparkSession

# Inicializar la sesión de Spark
spark = SparkSession.builder.appName("Datos_Tam").getOrCreate()

# Cargar el archivo CSV en un DataFrame de pandas
Datos_Tam = pd.read_csv("/content/TAM.csv")

# Convertir el DataFrame de pandas a un DataFrame de Spark
df_spark = spark.createDataFrame(Datos_Tam)

# Convertir el DataFrame de Spark a un RDD
rdd = df_spark.rdd

# Definir los valores a eliminar
valores_a_eliminar = [
    "Comercio al por menor en tiendas de abarrotes, ultramarinos y misceláneas",
    "Cafeterías, fuentes de sodas, neverías, refresquerías y similares",
    "Servicios de preparación de otros alimentos para consumo inmediato",
    "Comercio al por menor de frutas y verduras frescas",
    "Comercio al por menor de carnes rojas",
    "Comercio al por menor en tiendas de abarrotes",
    "Bares, cantinas y similares",
    "Comercio al por menor de otros alimentos",
    "Comercio al por menor de paletas de hielo y helados",
    "Comercio al por menor de bebidas no alcohólicas y hielo",
    "Comercio al por menor de pescados y mariscos",
    "Comercio al por menor de carne de aves",
    "Comercio al por menor de leche, otros productos lácteos y embutidos",
    "Centros nocturnos, discotecas y similares",
    "Servicios de preparación de alimentos en unidades móviles"
]

# Filtrar el RDD eliminando las filas con los valores de 'nombre_act' en la lista 'valores_a_eliminar'
# Accedemos a la columna 'nombre_act' usando su nombre
rdd_filtrado = rdd.filter(lambda x: x['nombre_act'] not in valores_a_eliminar)

# Verificar los primeros elementos del RDD filtrado (solo las primeras 5 filas)
print(rdd_filtrado.take(5))  # Esto imprimirá las filas del RDD filtrado

# Convertir el RDD filtrado de nuevo a un DataFrame de Spark (si es necesario)
df_filtrado = spark.createDataFrame(rdd_filtrado, df_spark.columns)

# Verificar el DataFrame filtrado
df_filtrado.show()
df_filtrado.count()
```

```python
#Aplicaremos otro filtro a nuestro df, ya que queremos que aprtir de la columna per_ocu nos muestre los valores a partir de 6 personas
# Filtrar el DataFrame para mostrar solo las filas donde per_ocu_num > 8
df_filtrado = df_spark.filter(df_spark.per_ocu_num >= 8)

# Mostrar el resultado
df_filtrado.show()
df_filtrado.count()
```
